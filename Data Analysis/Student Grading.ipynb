```python
import numpy as np
from numpy import random as R
import pandas as pd
import random as r
import csv
```

```python
#Q1 Read the given CSV file share with an assignment and use any python data structure
#   (DataFRames, NdArray etc) to load the data into the system.

df1=pd.read_csv('C:\\Users\\hp\\Desktop\\18BCE005\\2CS404_sessional_assignment.csv')
# pd.set_option('display.max_rows',None) 
df1
```

```python
#Q2 Fill the missing values of p6, p7, p8, p9 and p10. You can use any techniques to fill the missing
# value for all students.

p6=R.randint(0,11,size=df1['Rollno'].count())
p7=R.randint(0,11,size=df1['Rollno'].count())
p8=R.randint(0,11,size=df1['Rollno'].count())
p9=R.randint(0,11,size=df1['Rollno'].count())
p10=R.randint(0,11,size=df1['Rollno'].count())

temp=df1.iloc[:][:]
temp['Practical6']=p6
temp['Practical7']=p7
temp['Practical8']=p8
temp['Practical9']=p9
temp['Practical10']=p10
temp.drop(columns='ClassTestMark', axis=1, inplace=True)

# The temp dataframe will be used for all calculations related to LPW 
# Hence it does not contain ClassTestMark 
temp
```

```python
#Q3

x=df1.iloc[:,0:2]
sess=R.randint(0,41,size=df1['Rollno'].count())
ino=R.randint(0,31,size=df1['Rollno'].count())

x['SessionalMarks']=sess
x['Innovative_Assignment']=ino

a=x['ClassTestMark'].to_numpy()
x['Total']=a+sess+ino
p=np.where((a+sess+ino)>=40, 'Pass', 'Fail')
x['Pass/Fail']=p
csv1=x.iloc[:,0:len(x.columns)-1]
csv1.to_csv('C:\\Users\\hp\\Desktop\\18BCE005\\CE.csv',index=False,line_terminator='\n\n')

viva=R.randint(0,26,size=temp['Rollno'].count())
temp['Viva']=viva
pracs=temp.iloc[:,1:11].to_numpy()
sump=pracs.sum(axis=1)
total=sump*0.75 + viva
temp['Total']=total
d=np.where(total>=40, 'Pass', 'Fail')
temp['Pass/Fail']=d
csv2=temp.iloc[:,0:len(temp.columns)-1]
csv2.to_csv('C:\\Users\\hp\\Desktop\\18BCE005\\LPW.csv',index=False,line_terminator='\n\n')

see=pd.DataFrame(df1)
result=R.randint(0,101,size=see['Rollno'].count())
see['SEE_Marks']=result
p=np.where(result>=40,'Pass','Fail')
see['Pass/Fail']=p
see=see[['Rollno', 'SEE_Marks', 'Pass/Fail']]
csv3=see.iloc[:,0:len(see.columns)-1]
csv3.to_csv('C:\\Users\\hp\\Desktop\\18BCE005\\SEE.csv',index=False,line_terminator='\n\n')
```

```python
#Q5

final=pd.DataFrame(see)
sem=final['SEE_Marks'].to_numpy()
lpw=temp['Total'].to_numpy()
ce=x['Total'].to_numpy()
sem=sem*0.4
lpw=lpw*0.2
ce=ce*0.4
final['CE_Component']=ce
final['LPW_Component']=lpw
final['SEE']=sem
final['Grand_Total']=ce+lpw+sem
final=final[['Rollno','CE_Component','LPW_Component','SEE','Grand_Total']]

l=[]
for i in range(final['Rollno'].count()):
    pf='Failed in '
    s=''
    t=''
    d=''
    flag=0
    if x.loc[i]['Pass/Fail']=='Fail':
        s='CE'
        flag=1
    if temp.loc[i]['Pass/Fail']=='Fail':
        t='LPW'
        flag=1
    if see.loc[i]['Pass/Fail']=='Fail':
        d='SEE'
        flag=1
    if s!='':
        pf+=s+', '
    if t!='':
        pf+=t+', '
    if d!='':
        pf+=d+', '
    if pf[len(pf)-2 : len(pf)] == ', ':
        pf=pf[:len(pf)-2]
    if flag!=1:
        pf='Passed'
    l.append(pf)

l=np.array(l)
final['Remarks']=l
final.to_csv('C:\\Users\\hp\\Desktop\\18BCE005\\Marksheet.csv',index=False,line_terminator='\n\n')
final
```
