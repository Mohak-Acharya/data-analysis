```python
import csv,string,datetime 
import pandas as pd
import numpy as np
import numpy.random as r
import random
```

```python
#Q1 F1

def text(txtlen=8):
    letter = string.ascii_uppercase
    return ''.join(random.choice(letter) for i in range(txtlen))

n=50
#n=int(input('Enter number of employees'))
stdate = datetime.date(1950, 1, 1)
endate = datetime.date(2020, 1, 1)
d=(endate-stdate).days

row=['Employee_ID','Department_ID','Employee_Address','Employee_Phone','Employee_Type','Join_Date','Last_Promo'] 
etype=['Regular', 'Probation', 'Part time']

with open('C:\\Users\\hp\\Desktop\\F1.csv','w') as f1:
    w=csv.writer(f1)
    w.writerow(row)
    for i in range(50):
        l=[]
        eid=random.randrange(100000,1000000)
        l.append(eid)
        did=random.randrange(100,1000)
        l.append(did)
        add=text()
        l.append(add)
        phone=random.randrange(1000000000,10000000000)
        l.append(str(phone))
        typ=r.choice(etype)
        l.append(typ)
        joindate = stdate + datetime.timedelta(days=random.randrange(d))
        
        l.append(joindate.strftime('%d-%m-%Y'))
        if typ != 'Part time':
            day=(endate-joindate).days
            lastpromo=joindate + datetime.timedelta(days=random.randrange(day))
            l.append(lastpromo.strftime('%d-%m-%Y'))
        else:
            l.append('Nil')
        w.writerow(l)
```

```python
#Q2 F2
n=50
df=pd.read_csv('C:\\Users\\hp\\Desktop\\F1.csv')
row=['Employee_ID', 'Date', 'In_time', 'Out_time', 'Number_of_hours']
with open('C:\\Users\\hp\\Desktop\\F2.csv','w') as f2:
    w2=csv.writer(f2)
    w2.writerow(row)
    for i in range(n):
        l=[]
        eid=df.iloc[i,0]
        l.append(eid)
        s=datetime.datetime(datetime.datetime.now().year,datetime.datetime.now().month,datetime.datetime.now().day)
        l.append(s.strftime('%d-%m-%Y'))
        sec=random.randrange(60)
        mins=random.randrange(60)
        hr=random.randrange(7,12)
        t1=datetime.datetime(s.year,s.month,s.day,hr,mins,sec)
        l.append(t1.time().strftime("%I:%M:%S %p"))
        t2=t1+datetime.timedelta(hours=random.randrange(5,11),minutes=random.randrange(60),seconds=random.randrange(60))
        l.append(t2.time().strftime("%I:%M:%S %p"))
        x=(t2-t1+datetime.datetime.min).time()
        l.append(x.hour)
        w2.writerow(l)
```

```python
#Q3 F3
n=df['Department_ID'].count()
row=['Department_ID','Position','Payment_monthly','Payment_per_hour']
with open('C:\\Users\\hp\\Desktop\\F3.csv', 'w') as f3:
    w3=csv.writer(f3)
    w3.writerow(row)
    for i in range(n):
        l=[]
        l1=[]
        l2=[]
        l3=[]
        did=df.iloc[i,1]
        l1.append(did)
        l2.append(did)
        l3.append(did)
        l1.append('Part time')
        l2.append('Regular')
        l3.append('Probation')
        l1.append(np.nan)
        l1.append(random.randrange(500,1000,step=100))
        l2.append(random.randrange(20000,60001,step=1000))
        l2.append(np.nan)
        l3.append(random.randrange(25000,50001,step=1000))
        l3.append(np.nan)
        l.append(l1)
        l.append(l2)
        l.append(l3)
        w3.writerows(l)
```

```python
#Q4  Creating F4 using F1 F2 and F3

df1=pd.read_csv('C:\\Users\\hp\\Desktop\\F1.csv')
df2=pd.read_csv('C:\\Users\\hp\\Desktop\\F2.csv')
df3=pd.read_csv('C:\\Users\\hp\\Desktop\\F3.csv')

x=pd.merge(df1,df2,how='inner',on='Employee_ID')
x=x.loc[:,['Employee_ID','Department_ID','Employee_Type','Number_of_hours']]
y=pd.merge(x,df3,how='inner',left_on=['Department_ID','Employee_Type'],right_on=['Department_ID','Position'])
y=y.loc[:,['Employee_ID','Department_ID','Employee_Type','Number_of_hours','Payment_per_hour']]
y.drop_duplicates(subset='Employee_ID',keep='first',inplace=True)
y.reset_index(drop=True, inplace=True)
y.dropna(axis='index', how='any', inplace=True)
y.reset_index(drop=True, inplace=True)
p=y['Number_of_hours'].to_numpy()
q=y['Payment_per_hour'].to_numpy()
Amount_to_be_paid=p*q
y['Amount_to_be_paid']=Amount_to_be_paid
y=y[['Employee_ID', 'Number_of_hours', 'Amount_to_be_paid']]
t=datetime.datetime.now()
l=[t.strftime('%d-%m-%Y') for _ in range(y['Employee_ID'].count())]
l=np.array(l)
y.insert(0,'Date',l)
y.to_csv('C:\\Users\\hp\\Desktop\\F4.csv',line_terminator='\n\n',index=False)
y
```
